```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
iris_dataset = load_iris()
data_df = pd.DataFrame(iris_dataset['data'])
data_df.columns = iris_dataset['feature_names']
```

```python
data_df
```

```python
target_df = pd.DataFrame(iris_dataset['target'])
target_df
```

```python
target_names = pd.DataFrame(iris_dataset['target_names'])
target_names
```

```python
target_df[0][0]
```

```python
target_cat = pd.DataFrame(target_names[0][target_df[0]])
target_cat
```

```python
data_df['Target_Category'] = target_cat[0].values
data_df
```

```python
X = data_df.copy()
X.drop(columns='Target_Category', inplace=True)
```

```python
X
```

```python
Y = data_df[['Target_Category']]
Y
```

```python
data_df['Target_Category'].value_counts().sort_index().plot.bar()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.20)
```

```python
print("X_train size:" + str(X_train.shape))
print("Y_train size:" + str(Y_train.shape))
print("X_test size:" + str(X_test.shape))
print("Y_test size:" + str(Y_test.shape))
```

## KNN ALGORITHM:

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 2)
```

```python
Y_train
```

```python
np.array((Y_train)).ravel()
```

```python
knn.fit(X_train, np.array((Y_train)).ravel())
```

```python
Y_test
```

```python
Predictions = pd.DataFrame()
Predictions['Actual_y'] = Y_test['Target_Category'].copy()
Predictions
```

```python
Predictions = Predictions.reset_index()
Predictions
```

```python
Predictions.drop(columns=['index'], inplace=True)
Predictions
```

```python
Predictions['Predicted_y'] = pd.DataFrame(knn.predict(X_test)).copy()
Predictions
```

## Metrics

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
class_report = classification_report(Predictions['Predicted_y'], Predictions['Actual_y'])
print("Classification Report:")
print(class_report)
```

```python
#'setosa','versicolor','virginica'
confu_mat = pd.DataFrame(confusion_matrix(Predictions['Predicted_y'], Predictions['Actual_y']), 
                         index=['setosa','versicolor','virginica'], 
                         columns=['setosa','versicolor','virginica'])
confu_mat
```
